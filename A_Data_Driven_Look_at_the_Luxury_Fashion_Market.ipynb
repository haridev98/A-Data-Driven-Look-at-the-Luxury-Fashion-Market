### 3) Which category has the highest discount and how much?

```python
#calculating the discount percentage
main_df['discount_per'] = (main_df['price_us'] - main_df['price_us_dis'] )*100/main_df['price_us']
main_df['discount_per']
```

```python
sns.set_style('darkgrid')
sns.set_palette('Set2')
sns.set(rc={"figure.figsize":(7, 9)}) #width=7, height=9
sns.boxplot(x="discount_per", y="category_primary", data=main_df)
plt.title('Discount Box plot')
plt.xlabel('Discount')
plt.ylabel('Category')
sns.despine()
plt.show()
```

> **Inference** :The clothes category has the highest discount, almost 95 percent.


### 4) What is the most awarded discount rate?


```python
sns.set_style("whitegrid")
ax=sns.distplot(main_df.query('discount_per >0').discount_per, color="r")
ax.set(xlabel='Interest Rate %',
       ylabel='% Distribution',title='Density Plot of Interest Rate')

plt.legend();
```

> **Inference** : The most awarded discount rate is 60%, and it has been awarded to almost 10% of the products andthis discount is distributed across all category other than Shoes and Homeware which tends me to believe that there has been a masive discount campaign was ongoing with the farfetch to increase their sales.


### 5) What are the average rates of each product across all brands?

```python
# plotting Pivot table of product & brand for avg price
pivot_table = main_df.pivot_table(values='price_us_dis',index='category_primary',columns = "brand", aggfunc="mean")
pivot_table.fillna(0)
```

```python
# Making Dataframe for only values which are non zero.
brand =  sorted(set(main_df.brand.unique()))
temp_df = pd.DataFrame(columns =["category","val"] )
for b in brand:
  new_table = pivot_table[pivot_table[ b].notna()][ b]
  temp_df.loc[b] = {"category": list(new_table.index), "val": new_table.values.tolist()}

temp_df
```

```python
import os
import plotly.graph_objects as go

brand = sorted(set(main_df.brand.unique()[:30])) # limiting the list size due to computing resource constrain,
fig = go.Figure()

plot_names = []
buttons = []
default_state = 'OFF-WHITE'

for brand_name in brand:
  new_df = pd.DataFrame({'category':temp_df.loc[brand_name,'category'],'val':temp_df.loc[brand_name,'val']})
  fig.add_trace(go.Scatter(x=new_df['category'], y=new_df['val'], line={},mode='markers',marker=dict(size=40,symbol="star"), visible=(brand_name==default_state)))
  plot_names.extend([brand_name])

for brand_name in brand:
  index = brand.index(brand_name)  # Get the index of the brand
  visibility = [brand_name == name for name in plot_names]  # Create a list of visibility based on brand name
  buttons.append(dict(method='update',
                        label=brand_name,
                        args=[{'visible': visibility}])) # Update visibility based on brand name

fig.update_layout(showlegend=False, updatemenus=[{"buttons": buttons, "direction": "down", "active": brand.index(default_state), "showactive": True, "x": 0.8, "y": 1.50}])

fig.update_layout(template="plotly", title='<b>Average prices of Categories for all Brands</b>', xaxis_title="Category", yaxis_title="Avg Price ( $ )")
fig.update_layout()  # Add this line to update the layout

fig.show()
```

> **Inference** : The average product price varies across categories, and most brands sell products in a specific category. This suggests that these brands are specialized in that category.

### 6) Is there any product category have No discounts?

```python
def generate_chart(category , fig):
  df = main_df.query("category_primary == @category")
  count_values =main_df['discount'].value_counts()
  go.Figure(data=[go.Pie(values=count_values,labels=count_values.index, visible=(brand_name==default_state))])
  fig.add_trace( go.Figure(data=[go.Pie(values=count_values,labels=count_values.index)]))
  return fig
```

```python
#Add new column for legends in the PIE chart
main_df['discount_v']=main_df['discount'].map({1.0:"Discounted",0.0:"Full Price"})
```

```python
menu =sorted(set(main_df.category_primary.unique()))
fig = go.Figure()
plot_names = []
buttons = []
default_state = 'GIRLS CLOTHING'

for cat in menu:
  df = main_df.query("category_primary == @cat")
  count_values =df['discount_v'].value_counts()
  fig.add_trace( go.Pie(values=count_values,labels=count_values.index,
                        title =f'<b>Percentage of full price items and discounted items in category {cat}</b>',
                        visible=(cat==default_state)))
  plot_names.extend([cat])

for cat in menu:
  index = menu.index(cat)  # Get the index of the brand
  visibility = [cat == name for name in plot_names]  # Create a list of visibility based on brand name
  buttons.append(dict(method='update',
                        label=cat,
                        args=[{'visible': visibility}])) # Update visibility based on brand name

fig.update_layout(template="plotly",showlegend=True, updatemenus=[{"buttons": buttons, "direction": "down", "active": menu.index(default_state), "showactive": True, "x": 0.8, "y": 1.50}])
```

> **Inference** : Makeup, skincare, hair, and fragrance products are not typically discounted on Farfetch because they are seen as essential items and are often sold by luxury brands.

>Customers are more likely to purchase these products regardless of the price, so there is less need to offer discounts to attract sales. Additionally, luxury brands typically do not offer discounts on their products, as they want to maintain a high perceived value.

### 7) Specifically, how are products categorized based on their Category_primary, Category_seconday, and Category_tertiary?

```python
! pip install  wordcloud --quiet --upgrade
from PIL import Image
from wordcloud import WordCloud
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
def Word_Cloud(category , Color):
  text = main_df[category].values
  wordcloud = WordCloud( background_color=Color)
  wordcloud.generate(str(text))
  plt.imshow(wordcloud, interpolation='bilinear')
  plt.axis("off")
  plt.show()
```

```python
# WordCloud Of category_primary

Word_Cloud('category_primary','white')
```

```python
# WordCloud Of category_seconday

Word_Cloud('category_seconday','orange')
```

```python
# WordCloud Of category_tertiary

Word_Cloud('category_tertiary','yellow')
```

> **Inference** :The word clouds show that CLOTHING is the most common word in the category_primary, JACKETS is the most common word in the category_secondary, and BLAZERS is the most common word in the category_tertiary.
